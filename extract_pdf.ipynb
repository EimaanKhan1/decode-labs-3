```python
from PyPDF2 import PdfReader
reader = PdfReader('Cloud Computing Project 3.pdf')
for i, page in enumerate(reader.pages):
    page_text = page.extract_text() or ''
    print(f'--- page {i+1} ---')
    print(page_text[:4000])
    if i >= 1:
        break
print('Total pages extracted:', len(reader.pages))
```

```python
from PyPDF2 import PdfReader
reader = PdfReader('Cloud Computing Project 3.pdf')
for i, page in enumerate(reader.pages):
    page_text = page.extract_text() or ''
    clean = ' '.join(page_text.split())[:1000]
    print(f'--- page {i+1} summary ---')
    print(clean)
    print()
print('Total pages:', len(reader.pages))
```

```python
from PyPDF2 import PdfReader
reader = PdfReader('Cloud Computing Project 3.pdf')
keywords = [ 'requirement', 'requirements', 'task', 'tasks', 'deliverable', 'deliverables', 'objective', 'lab', 'project', 'due', 'submission', 'milestone']
for i, page in enumerate(reader.pages):
    text = page.extract_text() or ''
    clean = ' '.join(text.split())
    hits = [kw for kw in keywords if kw.lower() in clean.lower()]
    if hits:
        print(f'--- page {i+1} hits: {hits} ---')
        snippet = clean[:1200]
        print(snippet)
        print()
print('Total pages:', len(reader.pages))
```

```python
from PyPDF2 import PdfReader
reader = PdfReader('Cloud Computing Project 3.pdf')
with open('cloud_project_3_text.txt', 'w', encoding='utf-8') as f:
    for i, page in enumerate(reader.pages):
        text = page.extract_text() or ''
        f.write(f'--- PAGE {i+1} ---\n')
        f.write(text + '\n\n')
print('Wrote text to cloud_project_3_text.txt')
```
